```python
data = pd.read_csv("../input/wind.data")
date = data['Yr'].astype(str) + "-" + data['Mo'].astype(str) + "-" + data['Dy'].astype(str)
date_proper = pd.to_datetime(date, format='%y-%m-%d')
date_proper = pd.DataFrame(date_proper, columns =['Date_new'])
data = data.assign(Yr = date_proper)
data = data.rename(columns={'Yr':'Date_new'})
data = data.drop(columns=['Mo', 'Dy'])
data = data.set_index("Date_new")
data.head()
```

Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.

```python
# Minimum of the windspeeds across all locations for each week for the first 52 weeks
data.resample('W', on='Date_new').min().iloc[1:,0:12].head(52)
```

```python
# Maximum of the windspeeds across all locations for each week for the first 52 weeks
data.resample('W', on='Date_new').max().iloc[1:,0:12].head(52)
```

```python
# Mean of the windspeeds across all locations for each week for the first 52 weeks
data.resample('W', on='Date_new').mean().round(2).iloc[1:,0:12].head(52)
```

```python
# Standard Deviations of the windspeeds across all locations for each week for the first 52 weeks
data.resample('W', on='Date_new').std().round(2).iloc[1:,0:12].head(52)
```
