```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../input/occupation.csv", sep = '|')
```

```python
users = data.copy()

```

the mean age per occupation

```python

mean_age = users.groupby(by = ['occupation'])[['age']].mean().add_prefix('Mean_of_')
print(mean_age)
```

Discover the Male ratio per occupation and sort it from the most to the least.

Use numpy.where() to encode gender column.

```python
#users.head()

ratio = users.pivot_table(index='occupation', columns='gender', aggfunc='size', fill_value=0)

sums = ratio[['F', 'M']].sum(axis=1)

ratio['Male_Ratio'] = ratio['M'] / sums
ratio['Female_Ratio'] = ratio['F'] / sums
ratio.sort_values(by = ['Male_Ratio'],  ascending=False)
ratio.head()

```
